```python
farmers = pd.read_csv('farmers_dropna.csv')
farmers.head()
```

```python
farmers.drop(['long', 'lat'], axis=1)
farmers.head()
```

```python
features=[]
for i in range(farmers.shape[0]):
    x,y = farmers.latitude[i], farmers.longitude[i]
    latlong =[x,y]
    g=ee.Geometry.Point(latlong) 
    feature = ee.Feature(g)
    features.append(feature)
print("done")
```

```python
features[2].getInfo()
```

```python
type(features)
```

```python
farmer_features = ee.FeatureCollection(features)
```
